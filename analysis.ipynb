## Bad case analysis


```python
import matplotlib.pyplot as plt
import cv2
```

```python
best50 = rank_df.sort_values(by='rank')[:50]
best50
```

```python
worst50 = rank_df.sort_values(by='rank', ascending=False)[:50]
worst50
```

```python
########################
num = 8
# viz_df = best50.copy()
viz_df = worst50.copy()

k = 5
########################

image_id = viz_df.iloc[num]['ref_image']

fig, ax = plt.subplots(1, 2,  figsize=(4, 4))

img_path = f'fashionIQ_dataset/images/{image_id}.png'
img_ref = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)

target_id = viz_df.iloc[num]['captions']['target']
print(f"ref: {image_id}, target: {target_id}")
img_path = f'fashionIQ_dataset/images/{target_id}.png'
img_target = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)


ax[0].imshow(img_ref)
ax[0].set_title(f'ref {image_id}')

ax[1].imshow(img_target)
ax[1].set_title(f'target {target_id}')

print('captions: ', viz_df.iloc[num]['captions']['captions'])

fig1, ax1 = plt.subplots(1, k, figsize=(12, 12))
# target
top_preds_ids_row = shirt_preds_df[(shirt_preds_df['ref_image'] == image_id)&(shirt_preds_df[str(viz_df.iloc[num]['rank'])] == target_id)]
assert len(top_preds_ids_row) == 1, "not length 1"
top_preds_ids = top_preds_ids_row.iloc[0][0:k]
print('top 5 preds', top_preds_ids.tolist())
print('Label rank: ', viz_df.iloc[num]['rank'])
for i, id in enumerate(top_preds_ids):
    img_path = f'fashionIQ_dataset/images/{id}.png'
    img_topk = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
    ax1[i].imshow(img_topk)
    ax1[i].set_title(id, y=-0.01)
```
